```python
import tools, timeit, os
import pandas as pd
from astropy.io import fits
from astropy.utils.data import download_file
import matplotlib.pyplot as plt
from astropy.io import fits
import numpy as np
%config InlineBackend.rc = {}
%matplotlib inline
```

## Data information

```python
hdu_list = fits.open('./PTF1/img.fits')
```

```python
image1 = hdu_list[0].data
hdu_list.close()
```

```python
image1
```

```python
minpx = np.mean(image1)-1*np.std(image1)
maxpx = np.mean(image1)+1*np.std(image1)
plt.figure(figsize=(10,10))
plt.imshow(image1, origin='lower', cmap='gray')
plt.colorbar()
```

```python
hdu_list = fits.open('./PTF1/refimg.fits')
```

```python
image2 = hdu_list[0].data
hdu_list.close()
```

```python
minpx = np.mean(image2)-1*np.std(image2)
maxpx = np.mean(image2)+1*np.std(image2)
plt.figure(figsize=(10,10))
plt.imshow(image2, origin='lower', cmap='gray')
plt.colorbar()
```

```python
image3 = image1 - image2
```

```python
minpx = np.mean(image3)-2*np.std(image3)
maxpx = np.mean(image3)+2*np.std(image3)
plt.figure(figsize=(10,10))
plt.imshow(image3, origin='lower', cmap='gray')
plt.colorbar()
```

```python
image1conv = image1/np.std(image1)
plt.figure(figsize=(10,10))
plt.imshow(image1conv, origin='lower', cmap='gray')
plt.colorbar()
```

```python
image2conv = image2/np.std(image2)
plt.figure(figsize=(10,10))
plt.imshow(image2conv, origin='lower', cmap='gray')
plt.colorbar()
```

```python
image3conv = image1conv-image2conv
plt.figure(figsize=(10,10))
plt.imshow(image3conv, origin='lower', cmap='gray')
plt.colorbar()
```
